## Data Cleaning

```python
data.drop(['title',	'description',	'url', 'url_to_image', 'source_id'], inplace =True, axis=1)
data.head()
```

```python
data.drop(['content'], inplace=True, axis=1)
```

```python
new_data=data.copy()
```

```python
new_data.head()
```

```python
new_data.info()
```

```python
for index, row in new_data.iterrows():
    if pd.isnull(row['author']):
        new_data.at[index, 'author'] = row['published_at']        
```

```python
new_data.info()
```

```python
count=0
for index,row in new_data.iterrows():
    if row['source_name']=='The Irish Times' and row['top_article'] == 1:
        count+=1
print(count)
```

```python
top = new_data[new_data['top_article'] == 1]
mean = top[['engagement_reaction_count', 'engagement_comment_count', 
                            'engagement_share_count', 'engagement_comment_plugin_count']].mean()
for column in ['engagement_reaction_count', 'engagement_comment_count', 
               'engagement_share_count', 'engagement_comment_plugin_count']:
    new_data.loc[new_data['top_article'] == 1, column] = new_data.loc[new_data['top_article'] == 1, column].fillna(mean[column])
new_data.head()
```

```python
top = new_data[new_data['top_article'] == 1]
mean = new_data[['engagement_reaction_count', 'engagement_comment_count', 
                 'engagement_share_count', 'engagement_comment_plugin_count']].mean()

for column in ['engagement_reaction_count', 'engagement_comment_count', 
               'engagement_share_count', 'engagement_comment_plugin_count']:
    top.loc[:, column] = top.loc[:, column].fillna(mean[column])
new_data.update(top)
new_data.head()
```
