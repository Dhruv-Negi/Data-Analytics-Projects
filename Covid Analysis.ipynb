```python
data1.drop(['NewCases','NewDeaths','NewRecovered'],axis=1, inplace = True)
```

```python
data1.head(20)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Country/Region', y='TotalCases', data=data1.head(15),hue='TotalCases', palette='magma')
plt.xlabel('Country/Region')
plt.ylabel('Total Cases')
plt.title('Total cases in top 15 countries')
```

USAs High Case Count: The USA has a significantly higher number of total COVID-19 cases compared to other countries on the list.

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Country/Region', y='TotalCases', data=data1.head(15), hue='TotalDeaths', palette='magma')
plt.xlabel('Countries')
plt.ylabel('Total Cases')
plt.title('Total cases along with total deaths')
```

Disparity in Cases and Deaths: Theres a significant difference between the number of cases and deaths across countries, with the USA showing notably higher figures.

```python
plt.figure(figsize=(15,5))
sns.barplot(data=data1.head(15), x='Country/Region', y='TotalCases', hue='TotalRecovered', palette='magma')
plt.xlabel('Countries')
plt.ylabel('Total Recovered')
plt.title('Total Recovered along with total deaths')
```
