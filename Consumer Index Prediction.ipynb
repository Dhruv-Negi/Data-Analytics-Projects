```python
def fill_gaps(df, column):
    non_empty_indices = df[df[column].notna()].index
    for start, end in zip(non_empty_indices, non_empty_indices[1:]):
        if end - start > 1:
            mean_value = (df.loc[start, column] + df.loc[end, column]) / 2.0
            df.loc[start + 1:end - 1, column] = mean_value
    return df
fill_gaps(train_data,'NGDPRNSAXDCINQ')
```

```python
non_null_values = train_data['NGDPRNSAXDCINQ'].dropna()
differences = non_null_values.diff().dropna()
average_increase = differences.mean()

# Function to fill tail rows with the average increase
def fill_tail_with_average_increase(df, column, avg_increase):
    last_value = df[column].dropna().iloc[-1]
    null_indices = df[df[column].isna()].index
    for i, idx in enumerate(null_indices):
        df.at[idx, column] = last_value + (i + 1) * avg_increase
    return df

# Fill the tail rows with the average increase
fill_tail_with_average_increase(train_data, 'NGDPRNSAXDCINQ', average_increase)
```

```python
def fill_null_with_avg_increase(df, column):
    non_null_values = df[column].dropna()
    differences = non_null_values.diff().dropna()
    average_increase = differences.mean()
    
    # Fill null values at the end of the column
    last_value = non_null_values.iloc[-1]
    null_indices = df[df[column].isna()].index
    
    for i, idx in enumerate(null_indices):
        df.at[idx, column] = last_value + (i + 1) * average_increase
    
    return df

# Apply the function to all relevant columns
columns_to_fill = S_data.columns[1:]  # assuming 'ears' is not a numerical column to fill

for col in columns_to_fill:
    fill_null_with_avg_increase(S_data, col)
```

```python
def regressors(model_type, X_train, y_train, X_val, y_val):
    if model_type == 1:
        model = MultiOutputRegressor(Ridge(alpha=1.0))
    elif model_type == 2:
        model = MultiOutputRegressor(Lasso(alpha=0.1))
    elif model_type == 3:
        model = MultiOutputRegressor(RandomForestRegressor(random_state=42, n_jobs=-1, n_estimators=36, max_depth=38))
    elif model_type == 4:
        model = MultiOutputRegressor(GradientBoostingRegressor(min_samples_split=4, max_depth=6, n_estimators=160, learning_rate=0.4, random_state=42))
    elif model_type == 5:
        model = MultiOutputRegressor(XGBRegressor(booster='gbtree', min_child_weight=6, learning_rate=0.1, random_state=42, n_jobs=-1, n_estimators=140, max_depth=70))
    else:
        raise ValueError("Invalid model type. Choose from 1 (Ridge), 2 (Lasso), 3 (RandomForest), 4 (GradientBoosting), 5 (XGBoost).")

    model.fit(X_train, y_train)
    train_pred = model.predict(X_train)
    val_pred = model.predict(X_val)

    train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
    val_rmse = np.sqrt(mean_squared_error(y_val, val_pred))

    train_mae = np.mean(np.abs(y_train - train_pred))
    val_mae = np.mean(np.abs(y_val - val_pred))

    train_mse = mean_squared_error(y_train, train_pred)
    val_mse = mean_squared_error(y_val, val_pred)

    metrics = {
        'RMSE_train': train_rmse,
        'RMSE_val': val_rmse,
        'MAE_train': train_mae,
        'MAE_val': val_mae,
        'MSE_train': train_mse,
        'MSE_val': val_mse
    }

    return metrics
```
